# List Comprehensions

List comprehensions are used for creating new lists from other iterables.

As list comprehensions return lists, they consist of brackets containing the expression, which is executed for each element along with the for loop to iterate over each element.

```python
#example:
numbers = [1, 2, 3, 4]
squares = [n**2 for n in numbers]

print(squares)
```

```python
#access items; cannot access items by referring to an index
#example
thisset = {"apple", "banana", "cherry"}
for x in thisset:
  print(x)
```

Sets are immutable. Once created, we cannot change its contents.


```python
#adding items
thisset.add("orange")      #adding one item at a time
thisset.update(["orange", "mango", "grapes"])       #adding more than one item at a time.
```

```python
#removing items
thisset.remove("banana")
thisset.discard("banana")
x = thisset.pop()    #pop will remove only the last added element
thisset.clear()   #empties the set
del thisset   #delete the set completely
```

# Dictionary Operations

```python
thisdict= {"brand": "Ford",
  "model": "Mustang",
  "year": 1964}
#accessing items
x = thisdict["model"]  #or 
x = thisdict.get("model") #using get() method

#changing values
thisdict["year"] = 2018

#looping through dictionary
for x in thisdict:
  print(x)
#or
for x in thisdict.values():   #looping using the dict. values
  print(x)
#or
for key, value in thisdict.items():   #looping using key and value
  print(key, value)
```

```python
#check if a key exists
if "model" in thisdict:#using membership operator 'in'
  print("Yes, 'model' is one of the keys in the thisdict dictionary") 
```

```python
#adding items to dictionary
thisdict["color"] = "red"
print(thisdict)
```

```python
#removing items
thisdict.pop("model")  #using pop() method
print(thisdict)

thisdict.popitem()    #popitem() removes the last inserted item from the dictionary;in versions before 3.7 , it removes a random item
print(thisdict)

thisdict.clear()   #clear method clears the dictionary
print(thisdict)

#del can also delete the complete dictionary
del thisdict
```

# Dictionary methods
* clear()-	Removes all the elements from the dictionary

* copy()-	Returns a copy of the dictionary

* fromkeys()-	Returns a dictionary with the specified keys and value

* get()-	Returns the value of the specified key

* items()-	Returns a list containing a tuple for each key value pair

* keys()-	Returns a list containing the dictionary's keys

* pop()-	Removes the element with the specified key

* popitem()-	Removes the last inserted key-value pair

* setdefault()-	Returns the value of the specified key. If the key does not exist: insert the key, with the specified value

* update()-	Updates the dictionary with the specified key-value pairs

* values()-	Returns a list of all the values in the dictionary

# Membership and Identity operators

Membership operators are operators used to validate the membership of a value. It tests for membership in a sequence, such as strings, lists, or tuples.

```python
#'in' operator
list1=[1,2,3,4,5] 
list2=[6,7,8,9] 
for item in list1: 
    if item in list2: 
        print("overlapping")
```

```python
#'not in' operator
y = 20
list = [10, 20, 30, 40, 50 ]; 
  
if ( x not in list ): 
   print("x is NOT present in given list") 
```

Identity operator are used to determine whether a value is of a certain class or type. They are usually used to determine the type of data a certain variable contains.

```python
#'is' operator
x = 5
if (type(x) is int): 
    print("true") 
else: 
    print("false") 
```

```python
#'is not' operator
x = 5.2
if (type(x) is not int): 
    print("true") 
else: 
    print("false") 
```

# Lambda Functions

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

```python
#example
x = lambda a, b : a * b
print(x(5, 6))
```

```python
# Python3 program to
# demonstrate instantiating
# a class
class Car:
    # A simple class
    # attribute
    attr1 = "Petrol"
    attr2 = "750 HP"
    # A sample method 
    def start(self):
        print("Engine Started : Engine Type ", self.attr1)
        print("Ready to GO : Horse Power ", self.attr2)
# Driver code
# Object instantiation
BMW = Car()
# Accessing class attributes
# and method through objects
print(BMW.attr1)
BMW.start()
```
